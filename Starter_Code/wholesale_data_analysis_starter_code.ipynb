## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
subtotal = df["unit_price"] * df["qty"]
df["line_subtotal"] = subtotal
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def shipping_price(unit_weight) -> float:
    if unit_weight > 50:
            return unit_weight * 7 
    else:
            return unit_weight * 10

df["total_weight"] = df["unit_weight"] * df["qty"]
df["shipping_price"] = df["total_weight"].apply(shipping_price)
df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df["line_price"] = (df["line_subtotal"] + df["shipping_price"]) * 1.0925
df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df["line_cost"] = (df["unit_cost"] * df["qty"] + df["shipping_price"]) 
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
df["line_profit"] = (df["line_price"] - df["line_cost"]) 
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
client_list = [33615, 66037, 46820, 38378, 24741]
for client_id in client_list: 
    check_totals = round(df.loc[df["client_id"] == client_id, "line_price"].sum(), 2)
    print(f"{client_id} : ${check_totals}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
client_list = [33615, 66037, 46820, 38378, 24741]
qty_list = []
shipping_price_list = []
line_price_list = []
line_cost_list = []
line_profit_list = []
for client_id in client_list:
    total_units = round(df.loc[df["client_id"] == client_id, "qty"].sum(), 2)
    total_shipping = round(df.loc[df["client_id"] == client_id, "shipping_price"].sum(), 2)
    total_line_price = round(df.loc[df["client_id"] == client_id, "line_price"].sum(), 2)
    total_line_cost = round(df.loc[df["client_id"] == client_id, "line_cost"].sum(), 2)
    total_line_profit = round(df.loc[df["client_id"] == client_id, "line_profit"].sum(), 2)
    qty_list.append(total_units)
    shipping_price_list.append(total_shipping)
    line_price_list.append(total_line_price)
    line_cost_list.append(total_line_cost)
    line_profit_list.append(total_line_profit)

final_dict = {}
# qty	shipping_price	line_price	line_cost	line_profit

final_dict['client_id'] = client_list
final_dict['qty'] = qty_list
final_dict['shipping_price'] = shipping_price_list
final_dict['line_price'] = line_price_list
final_dict['line_cost'] = line_cost_list
final_dict['line_profit'] = line_profit_list

print(final_dict)

df_final = pd.concat({k: pd.Series(v) for k, v in final_dict.items()}, axis=1)

print(df_final)

```

```python
# Format the data and rename the columns to names suitable for presentation.
df_final.rename(columns={"client_id": "Client ID"}, inplace = True)
df_final.rename(columns={"qty": "Units"}, inplace = True)
df_final.rename(columns={"shipping_price": "Shipping (millions)"}, inplace = True)
df_final.rename(columns={"line_price": "Total Revenue (millions)"}, inplace = True)
df_final.rename(columns={"line_cost": "Total Cost (millions)"}, inplace = True)
df_final.rename(columns={"line_profit": "Total Profit (millions)"}, inplace = True)
display(df_final)


# Define the money columns. 
money_columns = ["Shipping (millions)", "Total Revenue (millions)", "Total Cost (millions)", "Total Profit (millions)"]

# Define a function that converts a dollar amount to millions.
def convert(amount):
    return amount / 1000000

# Apply the currency_format_millions function to only the money columns. 
df_final[money_columns] = df_final[money_columns].apply(convert)
(df_final)

# Rename the columns to reflect the change in the money format. 


```
